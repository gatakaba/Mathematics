# 単回帰

## 目的

## データ
- N : データの個数
- x : 年齢
- y : 給料

## パラメータ
- a : 切片
- b : 傾き
- sigma : 分散

## モデル

- $ Y [n] = y_{base}[n] + \epsilon_{n} $

- $y_{base} = a + b X[n]$

- $\epsilon \sim Normal(0,\sigma)$

```python
%matplotlib inline
```

```python
import pystan
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
stan_code = """
data {
    int < lower = 0 > N;
    real X[N];
    real Y[N];
}

parameters {
    real a;
    real b;
    real < lower = 0 > sigma;
}

model {
    for (n in 1:N){
        Y[n] ~ normal(a + b*X[n],sigma);
    }
}
"""
```

```python
df=pd.read_csv("data/data-salary.txt")

stan_dat = {
    'N': len(df),
    'X': df["X"],
    'Y': df["Y"]
}
fit = pystan.stan(model_code = stan_code, data = stan_dat, iter = 1000, chains = 4)
```

```python
print(fit)
```

```python
plt.rcParams['figure.figsize'] = (20, 12)
fit.plot()
plt.show()
```

```python
la = fit.extract(permuted=True)  # return a dictionary of arrays
plt.scatter(la["a"],la["b"])
```

```python
d=pd.DataFrame({'a':la['a'],'b':la['b'],'sigma':la['sigma']},columns=['a','b','sigma'])
```

```python
g = sns.pairplot(d)
```

```python
print(np.mean(d["a"]))
print(np.mean(d["b"]))

print(np.cov(np.c_[d["a"],d["b"]].T))
```

# 推定結果

```python
plt.scatter(df["X"],df["Y"])
x = np.linspace(20,65)
y = np.mean(d["a"]) + np.mean(d["b"])*x
plt.plot(x,y)
```

# 予測分布

```python
plt.scatter(df["X"],df["Y"])


l=[]
X=np.linspace(20,65,1000)

for x in X:
    l.append(d["a"] + d["b"]*x)

y=np.array(l)

plt.plot(X,np.median(y,axis=1))
plt.fill_between(X,np.mean(y,axis=1) - np.std(y,axis=1),np.mean(y,axis=1) + np.std(y,axis=1),alpha=0.5)
plt.fill_between(X,np.mean(y,axis=1) - np.std(y,axis=1)*3,np.mean(y,axis=1) + np.std(y,axis=1)*3,alpha=0.25)
```
