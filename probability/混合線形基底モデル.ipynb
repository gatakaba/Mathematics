```python
# coding:utf-8
import numpy as np
import matplotlib.pyplot as plt

mu1 = np.array([0, 10])

mu2 = np.array([30, 1])

cov1 = np.array([[1, 0],
                 [0, 1]])

cov2 = np.array([[1, 0],
                 [0,10]])

x = np.linspace(-5, 5, 1000)

y_list = []
for i in range(300):
    if np.random.random()<0.1:
        w = np.random.multivariate_normal(mu1, cov1)
        
        y=x * w[1] + w[0]
        plt.plot(x, y, c="r", alpha=0.25)
        
    else:
        w = np.random.multivariate_normal(mu2, cov2)
        y=x * w[1] + w[0]
        plt.plot(x, y, c="b", alpha=0.25)

        
```

## モデル

$p(s = 0) = \mu$

$p(s = 1) = 1 - \mu$

$p(w| s = 0) = N(w|m_{1},\Sigma_{1})$

$p(w| s = 1) = N(w|m_{2},\Sigma_{2})$


## 尤度関数

$p(t|w) = N(t|\mathbf{w}\phi(x),\beta^{-1})$

## 予測分布
$p(t|x) = \int p(t|w)p(w) dw$
