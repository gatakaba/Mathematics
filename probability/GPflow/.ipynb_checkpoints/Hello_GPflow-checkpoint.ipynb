# GP regression
GP regressionはGPflowの中で最も単純なモデルです。
潜在プロセスと雑音との共役のために、周辺尤度p(y|θ)は正確に計算することができます。

このノートでは、GPR(Gaussian Process Regression)モデルの構築方法と
パラメータθを最尤推定とMCMCによって求める方法を示します。

http://www.jmlr.org/papers/volume18/16-537/16-537.pdf

```python
import GPflow
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
```

サンプルデータを作成

```python
N = 200
X = np.random.rand(N,1)
Y = np.sin(12*X) + 0.66*np.cos(25*X) + np.random.randn(N,1)*0.1 + 3
plt.plot(X, Y, 'kx', mew=2)
```

## モデル構築
GPflowではモデルを作成するためにGPflowモデルクラスを使用します。このケースではGPRです。
カーネルkを作成し、生成されたデータとカーネルを使用して、GPRオブジェクトをインスタンス化します。
さらに尤度関数の分散を初期推定に設定します。

```python
k = GPflow.kernels.Matern52(1, lengthscales=0.3)
m = GPflow.gpr.GPR(X, Y, kern=k)
m.likelihood.variance = 0.01
```

## 予測
GPflowにはいくつかの予測方法があります.

```python
def plot(m):
    xx = np.linspace(-0.1, 1.1, 100)[:,None]
    mean, var = m.predict_y(xx)
    plt.figure(figsize=(12, 6))
    plt.plot(X, Y, 'kx', mew=2)
    plt.plot(xx, mean, 'b', lw=2)
    plt.fill_between(xx[:,0], mean[:,0] - 2*np.sqrt(var[:,0]), mean[:,0] + 2*np.sqrt(var[:,0]), color='blue', alpha=0.2)
    plt.xlim(-0.1, 1.1)
plot(m)
```

# 平均関数を使う
すべてのGPflowモデルはパラメータ化された平均関数を持つことができ、いくつかの単純なものはGPflow.mean_functionsで提供されます。

```python
k = GPflow.kernels.Matern52(1, lengthscales=0.3)
meanf = GPflow.mean_functions.Linear(1,0)
m = GPflow.gpr.GPR(X, Y, k, meanf)
m.likelihood.variance = 0.01
```

## パラメータの最適化
optimize()を用いることによって、簡単にMLの推定値を得ることができます。
デフォルトではL-BFGS-Bアルゴリズムによって求められます。

```python
m.optimize()
plot(m)
m
```

## MCMCを使う
MCMCによってGP regressionのハイパーパラメータ推定のためのデモを行う。
まず、カーネルパラメータを事前に設定し、次にMCMCを実行して、パラメータにどれくらい後方不確かさがあるのかを確認する。
まず適当な事前分布を設定する

```python
m.kern.lengthscales.prior=GPflow.priors.Gamma(1.,1.)
m.kern.variance.prior=GPflow.priors.Gamma(1.,1.)
m.likelihood.variance.prior=GPflow.priors.Gamma(1.,1.)
m.mean_function.A.prior=GPflow.priors.Gaussian(0.,10.)
m.mean_function.b.prior=GPflow.priors.Gaussian(0.,10.)
m
```

```python
samples=m.sample(500,epsilon=0.2,verbose=1)
```

```python
plt.plot(samples)
```
